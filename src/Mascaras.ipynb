```python
import cv2 as cv
import imutils
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
imagen = cv.imread('imagenes/onerice.png')
imgGray = cv.cvtColor(imagen, cv.COLOR_BGR2GRAY)
plt.imshow(imgGray, 'gray')
```

```python
def mascara(imagen):
    a, b = imagen.shape
    new_image = np.zeros((a,b))
    for i in range(0, a):
        for j in range(1, b - 1):
            new_image[i,j] = abs(imagen[i, j - 1] - imagen[i, j + 1])
    return new_image
```

```python
imagen_mascara = mascara(imgGray)
imagen_mascara = np.uint8(imagen_mascara)
plt.imshow(imagen_mascara, 'gray')
```

```python
histg = cv.calcHist([imagen_mascara],[0],None,[256],[0,256])
plt.plot(histg)
plt.show() 
```

```python
ret,thresh1 = cv.threshold(imagen_mascara,40,255,cv.THRESH_BINARY)
plt.imshow(thresh1, 'gray')
```

```python
histg = cv.calcHist([thresh1],[0],None,[256],[0,256])
plt.plot(histg)
plt.show() 
```
