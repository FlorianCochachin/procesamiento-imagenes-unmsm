```python
import cv2 as cv
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
ROOT_DIR = os.path.abspath(os.curdir)
DIR = os.path.join(ROOT_DIR, "imagenes")
os.chdir(DIR)
```

```python
img = cv.imread('frutos_rojos.png')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
print(img.shape)
```

```python
#Transformacion basica
#250 x 250
a, b, c = img.shape
a = int(a / 2)
b = int(b / 2)

img_new = np.zeros((a, b, c), int)

for i in range(a):
    for j in range(b):
        img_new[ i, j ] = img[ i * 2, j * 2 ]
plt.imshow(img_new)
plt.show()
```

```python
print(img_new.shape)
```

```python
#interpolacion
fx = 2.0
fy = 2.0
 # Expansin de interpolacin de vecino ms cercano
resized1 = cv.resize(img_new, dsize=None, fx=fx, fy=fy, interpolation = cv.INTER_NEAREST)
plt.imshow(resized1)
plt.show()
```

```python
#Traslacin
#io = f(i,j)
#jo = f(i,j)
#i0 = i + 10
#j0 = j + 20
#J
I = cv.imread('frutos_rojos.png',0)
a, b, c = img.shape
J = np.zeros((a, b))
#parametros de traslacin
for i in range(200, a):
   for j in range(120, b):
       J[i,j] = I[i,j]
plt.imshow(J)
plt.show()
```
