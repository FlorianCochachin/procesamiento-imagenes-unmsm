```python
import cv2 as cv
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv.imread('imagenes/reloj.jpg')
image = cv.cvtColor(image, cv.COLOR_BGR2RGB);
plt.imshow(image)
```

```python
histg = cv.calcHist([image],[0],None,[256],[0,256])
plt.plot(histg) 
plt.show() 
```

```python
def obtenerCanaleColor(image, color):
    a,b,h = image.shape
    image_new = np.zeros((a,b))
    for i in range(a):
        for j in range(b):
            image_new[i,j] = image[i,j, color]
    return np.uint8(image_new)
```

```python
image_new = obtenerCanaleColor(image, 2)
plt.imshow(image_new, 'gray')
```

```python
histg = cv.calcHist([image],[0],None,[256],[0,256])
plt.plot(histg) 
plt.show() 
```
