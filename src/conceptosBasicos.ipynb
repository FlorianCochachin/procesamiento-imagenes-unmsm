```python
import math
def obtenerMuestra(img, x, y, d):
    a, b = img.shape
    if d == 0:
        tam = 1
    else:
        tam = 3 + (d - 1) * 2
    muestra = np.zeros((tam,tam), int)
    ejex = range(x - d , x + d + 1)
    ejey = range(y - d, y + d + 1)
    for i in ejex:
        for j in ejey:
            muestra[abs(i - ejex[0]), abs(j - ejey[0])] = img[i,j]
    return muestra

imagen = obtenerMuestra(img, 10, 10, 2)

plt.imshow(imagen, 'gray')
plt.show()
```

```python
imagen_arroz = cv.imread('onerice.png',0)
plt.imshow(imagen_arroz, 'gray')
plt.show()
```

```python
line_ = imagen_arroz[160] # columna

plt.figure(figsize=(8, 6))
plt.plot(line_)

# Añadir etiquetas y título
plt.title('Grafica de una fila de la imágen')
plt.xlabel('W')
plt.ylabel('Pixel')

# Mostrar la gráfica
plt.show()
```

```python
a, h = imagen_arroz.shape
```

```python
new_image = np.zeros((a,h))

for i in range(a):
    for j in range(h):
        if imagen_arroz[i][j] > 180:
            new_image[i][j] = 1

```

```python
plt.imshow(new_image, 'gray')
plt.show()
```

```python
num_pixeles_1 = 0;
for i in range(a):
    for j in range(h):
        if new_image[i][j] == 1:
            num_pixeles_1 += 1

print("Area en pixeles:", num_pixeles_1)
        
```

```python
scale = cv.imread('scale.png', 0)
```

Cuantización (tonos de griz)

```python
p = 64

x,y = scale.shape
new_ = np.zeros((x,y), np.uint8)

for i in range(x):
    for j in range(y):
        v = int(np.fix(scale[i,j] / p) * p)
        new_[i,j] = v

print("Esta imagen tiene", int(256/p), " Tonosde gris")
plt.imshow(new_, 'gray')
plt.show()
```
