```python
import math
def obtenerMuestra(img, x, y, d):
    a, b = img.shape
    if d == 0:
        tam = 1
    else:
        tam = 3 + (d - 1) * 2
    muestra = np.zeros((tam,tam), int)
    ejex = range(x - d , x + d + 1)
    ejey = range(y - d, y + d + 1)
    for i in ejex:
        for j in ejey:
            muestra[abs(i - ejex[0]), abs(j - ejey[0])] = img[i,j]
    return muestra

imagen = obtenerMuestra(img, 400, 200, 30)

plt.imshow(imagen)
plt.show()
```
